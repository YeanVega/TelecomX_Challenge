```python
def rename_and_fix_dtypes(df: pd.DataFrame) -> pd.DataFrame:
    """
    Renombra columnas segÃºn el diccionario oficial y corrige tipos de datos
    clave para anÃ¡lisis de churn en Telecom X.
    """
    # 1. ---------------- Renombrar columnas -------------------------
    rename_map = {
        'customerID'                 : 'customerID',
        'Churn'                      : 'Churn',
        'customer.gender'            : 'gender',
        'customer.SeniorCitizen'     : 'SeniorCitizen',
        'customer.Partner'           : 'Partner',
        'customer.Dependents'        : 'Dependents',
        'customer.tenure'            : 'tenure',
        'phone.PhoneService'         : 'PhoneService',
        'phone.MultipleLines'        : 'MultipleLines',
        'internet.InternetService'   : 'InternetService',
        'internet.OnlineSecurity'    : 'OnlineSecurity',
        'internet.OnlineBackup'      : 'OnlineBackup',
        'internet.DeviceProtection'  : 'DeviceProtection',
        'internet.TechSupport'       : 'TechSupport',
        'internet.StreamingTV'       : 'StreamingTV',
        'internet.StreamingMovies'   : 'StreamingMovies',
        'account.Contract'           : 'Contract',
        'account.PaperlessBilling'   : 'PaperlessBilling',
        'account.PaymentMethod'      : 'PaymentMethod',
        'account.Charges.Monthly'    : 'Charges.Monthly',
        'account.Charges.Total'      : 'Charges.Total',
    }
    df = df.rename(columns=rename_map, errors="raise")

    # 2. ---------------- Tipos de datos -----------------------------
    # a) numÃ©ricos
    numeric_cols = ['tenure', 'Charges.Monthly', 'Charges.Total']
    df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')
    df[numeric_cols] = df[numeric_cols].astype('float32')

    # b) categÃ³ricos binarios â†’ strings coherentes en minÃºsculas
    bin_cols = [
        'Churn', 'gender', 'Partner', 'Dependents', 'PhoneService',
        'PaperlessBilling', 'SeniorCitizen'
    ]
    for col in bin_cols:
        df[col] = (
            df[col]      # respeta NaN si existen
            .astype(str)
            .str.strip()
            .str.lower()
            .replace({'nan': pd.NA})
        )

    return df
# -------------------------------------------------------------------
df = rename_and_fix_dtypes(df)
df.dtypes
```

## ðŸ”¹ IntroducciÃ³n

El objetivo principal de este anÃ¡lisis es **entender los factores asociados a la evasiÃ³n de clientes (Churn)** en TelecomX. La evasiÃ³n representa una preocupaciÃ³n clave en el sector de telecomunicaciones, ya que retener a un cliente suele ser mÃ¡s rentable que adquirir uno nuevo.

Este estudio permite identificar patrones de cancelaciÃ³n en funciÃ³n de caracterÃ­sticas contractuales, demogrÃ¡ficas y de facturaciÃ³n, aportando recomendaciones estratÃ©gicas basadas en evidencia.

---

## ðŸ”¹ Limpieza y Tratamiento de Datos

Los datos fueron cargados desde un archivo JSON (`TelecomX_Data.json`) y transformados en un DataFrame. Se realizaron los siguientes pasos:

* **Renombrado y correcciÃ³n de columnas**: Se eliminaron prefijos como `customer.`, `internet.`, `account.` para facilitar el anÃ¡lisis.
* **ConversiÃ³n de tipos**: Se transformÃ³ la columna `Charges.Total` a formato numÃ©rico (`float64`), ya que venÃ­a como `object`.
* **CÃ¡lculo de nueva variable**: Se generÃ³ la variable `Cuentas_Diarias` dividiendo `Charges.Monthly` entre 30 para estimar el valor de facturaciÃ³n diaria.
* **DetecciÃ³n de nulos**: Solo la variable `Charges.Total` presentÃ³ 11 valores nulos (0.15%), que fueron tratados adecuadamente.
* **Sin valores duplicados detectados**.

## ðŸ”¹ Conclusiones e Insights

* La **antigÃ¼edad** del cliente (tenure) es uno de los factores mÃ¡s determinantes para la retenciÃ³n. A mayor tiempo, menor probabilidad de churn.
* Los **clientes con contratos mensuales** son significativamente mÃ¡s propensos a cancelar.
* MÃ©todos de pago **automÃ¡ticos** reducen la evasiÃ³n. Electronic Check estÃ¡ asociado a la mayor tasa de cancelaciÃ³n.
* Los servicios de **fibra Ã³ptica**, aunque atractivos, podrÃ­an estar relacionados con una experiencia deficiente o costos mÃ¡s altos percibidos.
* El **gasto diario promedio es mayor en quienes cancelan**, lo que sugiere una percepciÃ³n de **alto costo sin valor percibido**.

1. **Fomentar contratos a largo plazo** mediante promociones, descuentos o beneficios exclusivos por renovaciÃ³n.
2. **Migrar clientes a mÃ©todos de pago automÃ¡ticos**, usando incentivos como descuentos o puntos de fidelizaciÃ³n.
3. **Evaluar la experiencia de usuarios de fibra Ã³ptica**: encuestas NPS o anÃ¡lisis de reclamos pueden revelar oportunidades de mejora.
