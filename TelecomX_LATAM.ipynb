```python
def rename_and_fix_dtypes(df: pd.DataFrame) -> pd.DataFrame:
    """
    Renombra columnas seg√∫n el diccionario oficial y corrige tipos de datos
    clave para an√°lisis de churn en Telecom X.
    """
    # 1. ---------------- Renombrar columnas -------------------------
    rename_map = {
        'customerID'                 : 'customerID',
        'Churn'                      : 'Churn',
        'customer.gender'            : 'gender',
        'customer.SeniorCitizen'     : 'SeniorCitizen',
        'customer.Partner'           : 'Partner',
        'customer.Dependents'        : 'Dependents',
        'customer.tenure'            : 'tenure',
        'phone.PhoneService'         : 'PhoneService',
        'phone.MultipleLines'        : 'MultipleLines',
        'internet.InternetService'   : 'InternetService',
        'internet.OnlineSecurity'    : 'OnlineSecurity',
        'internet.OnlineBackup'      : 'OnlineBackup',
        'internet.DeviceProtection'  : 'DeviceProtection',
        'internet.TechSupport'       : 'TechSupport',
        'internet.StreamingTV'       : 'StreamingTV',
        'internet.StreamingMovies'   : 'StreamingMovies',
        'account.Contract'           : 'Contract',
        'account.PaperlessBilling'   : 'PaperlessBilling',
        'account.PaymentMethod'      : 'PaymentMethod',
        'account.Charges.Monthly'    : 'Charges.Monthly',
        'account.Charges.Total'      : 'Charges.Total',
    }
    df = df.rename(columns=rename_map, errors="raise")

    # 2. ---------------- Tipos de datos -----------------------------
    # a) num√©ricos
    numeric_cols = ['tenure', 'Charges.Monthly', 'Charges.Total']
    df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')
    df[numeric_cols] = df[numeric_cols].astype('float32')

    # b) categ√≥ricos binarios ‚Üí strings coherentes en min√∫sculas
    bin_cols = [
        'Churn', 'gender', 'Partner', 'Dependents', 'PhoneService',
        'PaperlessBilling', 'SeniorCitizen'
    ]
    for col in bin_cols:
        df[col] = (
            df[col]      # respeta NaN si existen
            .astype(str)
            .str.strip()
            .str.lower()
            .replace({'nan': pd.NA})
        )

    return df
# -------------------------------------------------------------------
df = rename_and_fix_dtypes(df)
df.dtypes
```

# üìä Informe Final: An√°lisis de Evasi√≥n de Clientes (`Churn`) - TelecomX

---

## üîπ Introducci√≥n

El objetivo principal de este an√°lisis es **entender los factores asociados a la evasi√≥n de clientes (Churn)** en TelecomX. La evasi√≥n representa una preocupaci√≥n clave en el sector de telecomunicaciones, ya que retener a un cliente suele ser m√°s rentable que adquirir uno nuevo.

Este estudio permite identificar patrones de cancelaci√≥n en funci√≥n de caracter√≠sticas contractuales, demogr√°ficas y de facturaci√≥n, aportando recomendaciones estrat√©gicas basadas en evidencia.

---

## üîπ Limpieza y Tratamiento de Datos

Los datos fueron cargados desde un archivo JSON (`TelecomX_Data.json`) y transformados en un DataFrame. Se realizaron los siguientes pasos:

* **Renombrado y correcci√≥n de columnas**: Se eliminaron prefijos como `customer.`, `internet.`, `account.` para facilitar el an√°lisis.
* **Conversi√≥n de tipos**: Se transform√≥ la columna `Charges.Total` a formato num√©rico (`float64`), ya que ven√≠a como `object`.
* **C√°lculo de nueva variable**: Se gener√≥ la variable `Cuentas_Diarias` dividiendo `Charges.Monthly` entre 30 para estimar el valor de facturaci√≥n diaria.
* **Detecci√≥n de nulos**: Solo la variable `Charges.Total` present√≥ 11 valores nulos (0.15%), que fueron tratados adecuadamente.
* **Sin valores duplicados detectados**.

---

---

## üîπ Conclusiones e Insights

* La **antig√ºedad** del cliente (tenure) es uno de los factores m√°s determinantes para la retenci√≥n. A mayor tiempo, menor probabilidad de churn.
* Los **clientes con contratos mensuales** son significativamente m√°s propensos a cancelar.
* M√©todos de pago **autom√°ticos** reducen la evasi√≥n. Electronic Check est√° asociado a la mayor tasa de cancelaci√≥n.
* Los servicios de **fibra √≥ptica**, aunque atractivos, podr√≠an estar relacionados con una experiencia deficiente o costos m√°s altos percibidos.
* El **gasto diario promedio es mayor en quienes cancelan**, lo que sugiere una percepci√≥n de **alto costo sin valor percibido**.

---

## üîπ Recomendaciones Estrat√©gicas

1. **Fomentar contratos a largo plazo** mediante promociones, descuentos o beneficios exclusivos por renovaci√≥n.
2. **Migrar clientes a m√©todos de pago autom√°ticos**, usando incentivos como descuentos o puntos de fidelizaci√≥n.
3. **Evaluar la experiencia de usuarios de fibra √≥ptica**: encuestas NPS o an√°lisis de reclamos pueden revelar oportunidades de mejora.
4. **Segmentar clientes con bajo tenure y alta facturaci√≥n diaria** para aplicar estrategias de retenci√≥n anticipadas.
5. **Fortalecer el v√≠nculo con clientes sin pareja o dependientes**, quienes presentan mayor churn, mediante programas de fidelizaci√≥n.
